```python
import os
import cv2
import numpy as np
import segmentation_models as sm
import random
import glob
import keras
import re
import keras.backend as K
import matplotlib.pyplot as plt
from segmentation_models import Unet
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.layers import Input
from keras.models import Model
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
train_size = 0.6
val_size = 0.2
test_size = 0.2

data_dir = '/content/data_wo_shadows'
image_dir = os.path.join(data_dir, 'images_wo_gripper')
mask_dir = os.path.join(data_dir, 'segmentation_masks_wo_gripper')

# Get a list of all image filenames in the image_dir
all_image_filenames = os.listdir(image_dir)
selected_image_filenames = random.sample(all_image_filenames, 5000)

# create a list of mask filenames that correspond to the selected image filenames
selected_mask_paths = []
filtered_image_filenames = []
skipped_image_paths = []
for image_filename in selected_image_filenames:
    image_number = os.path.splitext(image_filename)[0]
    image_number = "".join(re.findall(r'\d+', image_number))
    mask_filename = image_number + '.png'
    mask_path = os.path.join(mask_dir, mask_filename)
    
    if os.path.exists(mask_path):
          selected_mask_paths.append(mask_path)
          filtered_image_filenames.append(image_filename)
    else:
        skipped_image_paths.append(os.path.join(image_dir, image_filename))
        

# create lists of image and mask paths for the selected images
selected_image_paths = [os.path.join(image_dir, f)
                        for f in filtered_image_filenames]

#keep green instance
def keep_green_instance(selected_mask_paths):
    mask = cv2.imread(selected_mask_paths, cv2.IMREAD_COLOR)
    lower_green = np.array([0, 100, 0])
    upper_green = np.array([100, 255, 100])
    green_mask = cv2.inRange(mask, lower_green, upper_green)
    return green_mask

image_dataset = []
mask_dataset = []
for img_path, mask_path in zip(selected_image_paths, selected_mask_paths):
    # load the image
    img = cv2.imread(img_path, cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)        

    # load the mask
    mask = keep_green_instance(mask_path)  
    
    image_dataset.append(img)
    mask_dataset.append(mask)

image_dataset = np.array(image_dataset)
mask_dataset = np.array(mask_dataset)

# expand dimensions for iamges and masks
mask_dataset = np.expand_dims(mask_dataset, axis=-1)


# image_dataset = image_dataset / 255.
# mask_dataset = mask_dataset / 255.

image_dataset = image_dataset.astype(np.float32) / 255.
mask_dataset = mask_dataset.astype(np.float32) / 255.


#x_train, x_val, y_train, y_val = train_test_split(image_dataset, mask_dataset, test_size=0.2, random_state=42)
# First, split the data into train and a temporary set (val + test)
x_train, temp_paths, y_train, temp_mask_paths = train_test_split(
    image_dataset, mask_dataset, train_size=train_size, random_state=42)

# Next, split the temporary set into validation and test sets
val_size_ratio = val_size / (val_size + test_size)
x_val, x_test, y_val, y_test = train_test_split(
    temp_paths, temp_mask_paths, test_size=1-val_size_ratio, random_state=42)
```

```python
from keras.optimizers import Adam

# backbone
BACKBONE = 'resnet101'
preprocess_input = sm.get_preprocessing(BACKBONE)

# define model
base_model = sm.Unet(
    BACKBONE, 
    classes=1, 
    activation='sigmoid',
    encoder_weights='imagenet',
    encoder_freeze=True
)

# create model with preprocessing layers
input = Input(shape=(480, 480, 3), name='Input')
preprocessed = preprocess_input(input)
output = base_model(preprocessed)
# final model
model = Model(input, output, name='BaseModel')

model.compile(
    #'Adam',
    Adam(learning_rate=1e-5),
    #loss=sm.losses.bce_jaccard_loss,
    loss=sm.losses.dice_loss,
    #loss=sm.losses.binary_focal_loss,
    #loss='binary_crossentropy',
    metrics=[sm.metrics.iou_score],
)

early_stopping_callback = EarlyStopping(
    monitor='val_loss', patience=3, verbose=1)

checkpoint_callback = ModelCheckpoint(
    filepath='best_model_weights.h5',
    monitor='val_loss',
    save_weights_only=True,
    save_best_only=True,
    verbose=1
)
```

```python
# fit the model
history = model.fit(
    x=x_train,
    y=y_train,
    batch_size=16,
    epochs=100,  # Increase the number of epochs
    validation_data=(x_val, y_val),
    callbacks=[checkpoint_callback, early_stopping_callback]
)
```
