```python
import matplotlib.pyplot as plt
import os
import cv2
import numpy as np
import random
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

```python
# Function to compute angle between three points
def compute_angle(pt1, pt2, pt3):
    a = np.array(pt1)
    b = np.array(pt2)
    c = np.array(pt3)

    ba = a - b
    bc = c - b

    cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))
    angle = np.arccos(cosine_angle)

    return np.degrees(angle)

# Load the mask paths
mask_paths = ['/content/drive/MyDrive/predicted_masks3/mask_23.png', '/content/drive/MyDrive/predicted_masks3/mask_12.png', '/content/drive/MyDrive/predicted_masks3/mask_25.png', '/content/drive/MyDrive/predicted_masks3/mask_15.png', '/content/drive/MyDrive/predicted_masks3/mask_53.png',  '/content/drive/MyDrive/predicted_masks3/mask_55.png', '/content/drive/MyDrive/predicted_masks3/mask_63.png', '/content/drive/MyDrive/predicted_masks3/mask_80.png' ]

for mask_path in mask_paths:

    # Load the image in color
    image = cv2.imread(mask_path)

    # Convert the image to grayscale for processing
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Find all the contours in the image
    contours, _ = cv2.findContours(gray, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    # Draw all contours on the image
    cv2.drawContours(image, contours, -1, (0, 255, 0), 2)
    # Loop over the contours
    for contour in contours:

        # Create an empty list to store the grasping points
        grasping_points = []

        # Compute the convex hull of the contour
        hull = cv2.convexHull(contour, returnPoints = False)

        # Compute the convexity defects
        defects = cv2.convexityDefects(contour, hull)
        if defects is None:
          continue
        # Loop over the convexity defects
        for i in range(defects.shape[0]):

            s,e,f,d = defects[i,0]
            start = tuple(contour[s][0])
            end = tuple(contour[e][0])
            far = tuple(contour[f][0])

            # Compute the angle formed at the defect point
            angle = compute_angle(start, far, end)

            # If the angle is wide (close to 180 degrees), consider it as a potential grasping point
            if angle > 130 and angle < 180:
                grasping_points.append(far)

        # Convert the list of grasping points to a numpy array
        grasping_points = np.array(grasping_points)

        # Apply DBSCAN to cluster the grasping points
        clustering = DBSCAN(eps=38, min_samples=1).fit(grasping_points)

        # Get the labels assigned to each point
        labels = clustering.labels_

        # Initialize an empty list to store the final grasping points
        final_grasping_points = []

        # Loop over each unique label
        for label in set(labels):
            if label != -1:
                # Get the grasping points that belong to the current cluster
                cluster_points = grasping_points[labels == label]

                # Compute the mean of the cluster points
                cluster_mean = np.mean(cluster_points, axis=0).astype(int)

                # Append the mean point to the final graspingpoints list
            final_grasping_points.append(tuple(cluster_mean))

        # Draw the final grasping points on the image
        for point in final_grasping_points:
            cv2.circle(image, tuple(point), 7, (0, 0, 255), -1)

    # Display the image
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Detected Handles and Grasping Points')
    plt.show()
```
