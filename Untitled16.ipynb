```python
# po folderima bez redundancije
import os
import cv2
import numpy as np
import segmentation_models as sm
import random
import glob
import keras
import re
import keras.backend as K
import matplotlib.pyplot as plt
from segmentation_models import Unet
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

BACKBONE = 'resnet34'
preprocess_input = sm.get_preprocessing(BACKBONE)
keras.backend.set_image_data_format('channels_last')

data_dir = '/content/data'
image_dir = os.path.join(data_dir, 'images')
input_shape = (480, 480, 3)

# specify the list of mask filenames you want to extract
mask_filenames = []
for i in range(6000):
    filename = 'LargeAmmoCanHandles.001_frame_{:04d}.png'.format(i)
    mask_filenames.append(filename)

# Get a list of all image filenames in the image_dir
all_image_filenames = os.listdir(image_dir)

# randomly select 100 image filenames
selected_image_filenames = random.sample(all_image_filenames, 150)

# create a list of mask filenames that correspond to the selected image filenames
selected_mask_paths = []
filtered_image_filenames = []
skipped_image_paths = []
for image_filename in selected_image_filenames:
    image_number = os.path.splitext(image_filename)[0]
    image_number = "".join(re.findall(r'\d+', image_number))
    mask_path = os.path.join(data_dir, 'masks', image_number)
    mask_found = False
    for mask_name in mask_filenames:
        current_mask_path = os.path.join(mask_path, mask_name)
        if os.path.exists(current_mask_path):
            mask_found = True
            selected_mask_paths.append(current_mask_path)
            break
    if mask_found:
        filtered_image_filenames.append(image_filename)
    else:
        print("No mask file found for selected image: {}. Skipping image...".format(image_filename))
        skipped_image_paths.append(os.path.join(image_dir, image_filename))

def visualize_skipped_images(images, num_examples=10):
    fig, axes = plt.subplots(num_examples, 1, figsize=(5, num_examples * 5))

    for i in range(num_examples):
        img = cv2.imread(images[i], cv2.IMREAD_COLOR)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        image_name = os.path.basename(images[i])  # Extract the image name from the path
        axes[i].imshow(img)
        axes[i].set_title("Skipped Image {}".format(image_name))

    plt.tight_layout()
    plt.show()

# Visualize skipped images
num_skipped_images_to_show = 5  # Or any number you'd like to show
visualize_skipped_images(skipped_image_paths, num_skipped_images_to_show)


# create lists of image and mask paths for the selected images
selected_image_paths = [os.path.join(image_dir, f) for f in filtered_image_filenames]

print(selected_image_paths)
print(selected_mask_paths)

# split the selected image and mask paths into train and validation sets
train_paths, val_paths, train_mask_paths, val_mask_paths = train_test_split(selected_image_paths, selected_mask_paths, test_size=0.2, random_state=42)

# load the train images and masks
x_train = []
y_train = []
# load the validation images and masks
x_val = []
y_val = []

all_paths = train_paths + val_paths
all_mask_paths = train_mask_paths + val_mask_paths

x_data = []
y_data = []
for img_path, mask_path in zip(all_paths, all_mask_paths):
    # load the image
    img = cv2.imread(img_path, cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, input_shape[:2], interpolation=cv2.INTER_LINEAR)

    img = img / 255.0
    # # preprocess the image
    img = preprocess_input(img)
    

    # load the mask
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    mask = cv2.resize(mask, input_shape[:2], interpolation=cv2.INTER_NEAREST)

    # threshold the mask to convert to binary mask
    _, mask = cv2.threshold(mask, 127, 255, cv2.THRESH_BINARY)

    # apply morphological opening operation to remove small noise in the mask
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

    # add to the lists
    x_data.append(img)
    y_data.append(mask)

# Separate the combined lists into train and validation lists
x_train = x_data[:len(train_paths)]
y_train = y_data[:len(train_mask_paths)]
x_val = x_data[len(train_paths):]
y_val = y_data[len(train_mask_paths):]

# convert to numpy arrays
x_train = np.array(x_train)
y_train = np.array(y_train)

# convert to numpy arrays
x_val = np.array(x_val)
y_val = np.array(y_val)

def load_data():
    # preprocess input
    x_train_pre = (x_train)
    x_val_pre = (x_val)

    # resize masks to fixed size
    mask_size = (480, 480) # set a fixed size for the masks
    y_train_resized = [cv2.resize(mask, mask_size, interpolation=cv2.INTER_NEAREST) for mask in y_train]
    y_val_resized = [cv2.resize(mask, mask_size, interpolation=cv2.INTER_NEAREST) for mask in y_val]

    # convert the lists to numpy arrays
    y_train_resized = np.array(y_train_resized)
    y_val_resized = np.array(y_val_resized)

    # normalize masks
    y_train_normalized = y_train_resized / 255.0
    y_val_normalized = y_val_resized / 255.0

    
    # expand dimensions for masks
    y_train_expanded = np.expand_dims(y_train_normalized, axis=-1)
    y_val_expanded = np.expand_dims(y_val_normalized, axis=-1)

    return x_train_pre, y_train_expanded, x_val_pre, y_val_expanded


# define model
model = sm.Unet(BACKBONE, classes=1, activation='sigmoid', encoder_weights='imagenet')
model.compile(
    'Adam',
    loss=sm.losses.bce_jaccard_loss,
    metrics=[sm.metrics.iou_score],
)

# load your data
x_train, y_train, x_val, y_val = load_data()

def visualize_images_and_masks_raw(image_paths, mask_paths, num_examples=10):
    fig, axes = plt.subplots(num_examples, 2, figsize=(10, num_examples * 5))

    for i in range(num_examples):
        # Read the image and mask from the paths
        image = cv2.imread(image_paths[i])
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        mask = cv2.imread(mask_paths[i], cv2.IMREAD_GRAYSCALE)

        axes[i, 0].imshow(image)
        axes[i, 0].set_title("Image: {}".format(os.path.basename(image_paths[i])))
        axes[i, 1].imshow(mask, cmap='gray')
        axes[i, 1].set_title("Mask: {}".format(os.path.basename(mask_paths[i])))

    plt.tight_layout()
    plt.show()

visualize_images_and_masks_raw(train_paths, train_mask_paths)
visualize_images_and_masks_raw(val_paths, val_mask_paths)

early_stopping_callback = EarlyStopping(monitor='val_loss', patience=15, verbose=1)

checkpoint_callback = ModelCheckpoint(
    filepath='model_weights_ep{epoch:02d}.h5',
    monitor='val_loss',
    save_weights_only=True,
    save_best_only=False,
    verbose=1
)

# fit the model
history = model.fit(
   x=x_train,
   y=y_train,
   batch_size=16,
   epochs=50,  # Increase the number of epochs
   validation_data=(x_val, y_val),
   callbacks=[checkpoint_callback, early_stopping_callback]
)
```

```python
import matplotlib.pyplot as plt

# Plot training & validation IoU scores
plt.figure(figsize=(12, 6))
plt.plot(history.history['iou_score'], label='Training IoU Score')
plt.plot(history.history['val_iou_score'], label='Validation IoU Score')
plt.title('Model IoU Score')
plt.ylabel('IoU Score')
plt.xlabel('Epoch')
plt.legend(loc='upper left')
plt.show()

# Plot training & validation loss values
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper right')
plt.show()
```
